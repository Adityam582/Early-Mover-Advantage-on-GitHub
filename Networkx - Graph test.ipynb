```python
import networkx as nx  
import matplotlib.pyplot as plt 

#create a new graph
G = nx.Graph()

#add edges
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (2, 5), (3, 4), (4, 5), (4, 6), (5, 7), (5, 8), (7, 8)])
```

```python

#size of the graph
plt.figure(figsize =(9, 12)) 

# original Graph created
# nrows=2, ncols=1, plot_number=1
plt.subplot(211) 

#the shape changes with every execution
nx.draw_networkx(G)
```

```python

#different layout, show name of nodes, this keeps the same shape
nx.draw_kamada_kawai(G,with_labels=True)

#to save image as file
#plt.savefig("graph.png")
```

```python
#different layout, this keeps the same shape
nx.draw_circular(G,with_labels=True)
```

```python
#different layout, this keeps the same shape
nx.draw_spectral(G,with_labels=True)
```

```python
#different layout, shape of graph changes with every execution
nx.draw_spring(G, with_labels=True)
```

```python
#different layout, this keeps the same shape
nx.draw_shell(G, with_labels=True)
```

```python
#another graph
g2 = nx.Graph()

list_of_edges = [(0,3), (3,5),(3,1),(3,2),(3,4),(5,1),(1,2),(2,4)]


#add edges
g2.add_edges_from(list_of_edges)
```

```python
nx.draw_spectral(g2,with_labels=True)
```

### Degree centrality for each node, calculate node with highest degree centrality

```python
#returns a dictionary with:
# node-label : centrality degree

max_degree_centrality = nx.degree_centrality(g2)

print (max_degree_centrality)

#print the node that has the highest centrality degree
print ( "\n", max(max_degree_centrality, key=max_degree_centrality.get) )
```

### Eigenvector centrality for each node, calculate node with highest eigenvector centrality

```python
#returns a dictionary with:
# node-label : eigenvector centrality degree


eigenv_centrality= nx.eigenvector_centrality(g2)

print(eigenv_centrality)

#print the node that has the highest eigen centrality degree
print ( "\n", max(eigenv_centrality, key=eigenv_centrality.get) )
```

### Betweenes centrality
####  sum of the fraction of all-pairs shortest paths that pass through a node

```python
#returns a dictionary with:
# node-label : betweenes centrality degree
# normalize, does the adjust automatically (directed and undirected graphs)
btw_centrality = nx.betweenness_centrality(g2, normalized=True)

print(btw_centrality)

#print the node that has the highest betweenes centrality value
print ( "\n", max(btw_centrality, key=btw_centrality.get) )
```

### Closeness centrality
#### Average shortest path distance to a node u, over all n-1 reachable nodes

```python
#returns a dictionary with:
# node-label : betweenes centrality degree

closeness_c = nx.closeness_centrality(g2)

print(closeness_c)

#print the node that has the highest closeness centrality value
print ( "\n", max(closeness_c, key=closeness_c.get) )
```

### Print a part of a graph, indicating which nodes to consider ( print subgraph)

```python
#sub graph of graph G (top part of code)
# [1, 2, 3, 4] are nodes in the original set of nodes 

sub_G = G.subgraph([1, 2, 3, 4])

  
plt.subplot(211) 

nx.draw_networkx(sub_G) 
```
